Goal: Predict which flower species an iris belongd to.

classes:
1.Setosa 2.Versicolor 3.Virginica

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow import argmax #converts probability output -> class index
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
#load the iris dataset
iris = load_iris()
x = iris.data
y = iris.target
target_names = iris.target_names
print(target_names)
```

```python
#plot the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
#standardize the features
scalar = StandardScaler()
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.transform(x_test)
```

```python
#Define the model archihtecture
model = Sequential()
model.add(Dense(10, activation='relu', input_shape = (x_train.shape[1],)))
model.add(Dense(5, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(x_train_scaled, y_train, epochs = 70, validation_split = 0.1, verbose=1)
```

```python
test_loss, test_accuracy = model.evaluate(x_test_scaled, y_test)
print(f"Test accuracy: ", test_accuracy)
```

```python
#make predictions
y_pred_proba = model.predict(x_test_scaled) #[[0.98 0.01 0.01][0.02 0.94 0.04]]
y_pred = argmax(y_pred_proba, axis = 1)

#calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy}")
```
