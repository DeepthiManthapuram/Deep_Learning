# **XOR Problem using Multi-layer(Forward propagation)**
Architecture:
2 inputs -> 2 hidden neurons -> 1 output

```python
import numpy as np
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
x, y
```

**Initialise weigths and biases(Manually chosen**

```python
#hidden layer weigths and biases
w1, w2, b1 = 1, 1, -0.5
w3, w4, b2 = 1, 1, -1.5

#output layer weigths and bisa
w5, w6, b3 = 1, -2, -0.5

print("weigths initialised successfully!")


```

**Define step activation function**

```python
def step_function(z):
  return 1 if z >= 0 else 0
```

**Define froward propagation function**

```python
def forward_propagation(x1, x2):
  #hidden layer calculation
  h1 = w1 * x1 + w2 * x2 + b1
  a_h1 = step_function(h1)

  h2 = w3 * x1 + w4 * x2 + b2
  a_h2 = step_function(h2)

  y = w5 * a_h1 + w6 * a_h2 + b3
  output = step_function(y)

  return output
```

**Test Forward Propagation for all XOR inputs**

```python
print("XOR Results using ANN Forward Propagation")

for input_row in x:
  x1, x2 = input_row[0], input_row[1]
  prediction = forward_propagation(x1, x2)
  print(f"Input: {input_row} Output: {prediction}")
```
