# **Perceptron(Binary Classification)**
EX: study hours vs pass/fail

```python
import numpy as np
```

**1.Prepare training data**

```python
#training data
x = np.array([1, 2, 3, 4]) #study hours
y = np.array([0, 0, 1, 1]) #pass(1)/fail(0)
x, y
```

**2.Initialize parameters**

```python
w = 0.0
b = 0.0
learning_rate = 1
epochs = 5

print("Initial weight: ", w)
print("Initial bias: ", b)
```

**3.Define Activation Function (Step function)**

```python
def step_function(z):
  return 1 if z >= 0 else 0
```

**Training loop**

```python

for epoch in range(epochs):
  print(f"\nEpoch {epoch + 1}")
  for i in range(len(x)):
    #1. compute weighted sum
    z = w * x[i] + b

    #2. Apply activation function
    y_pred = step_function(z)

    #3. compute error
    error = y[i] - y_pred

    #4. update weight
    w = w + learning_rate * error * x[i]

    #5. update bias
    b = b + learning_rate * error

    print(f"Input: {x[i]}, Predicted: {y_pred}, Actual: {y[i]}, w: {w}, b: {b}" )
```

**Final model parameters**

```python
print("\n Final weight: ",w)
print("\n Final bias: ",b)
```

**Testing the model**

```python
test_value = 2.9
z = w * test_value + b
prediction = step_function(z)

print(f"for study hours {test_value}, Prediction: ", prediction)
```

```python
import matplotlib.pyplot as plt

# Calculate the decision boundary
# The decision boundary is where w * x + b = 0, so x = -b / w
x_decision_boundary = -b / w

# Create a range of x values for plotting the step function
x_plot = np.linspace(min(x) - 1, max(x) + 1, 100)
y_plot = [step_function(w * val + b) for val in x_plot]

plt.figure(figsize=(8, 6))

# Plot training data
plt.scatter(x, y, color=['blue' if val == 0 else 'red' for val in y], label='Training Data')

# Plot the perceptron's output (step function)
plt.plot(x_plot, y_plot, linestyle='--', color='green', label='Perceptron Output')

# Plot the decision boundary
plt.axvline(x=x_decision_boundary, color='purple', linestyle=':', label=f'Decision Boundary (x = {x_decision_boundary:.2f})')

plt.title('Perceptron Decision Boundary')
plt.xlabel('Study Hours (x)')
plt.ylabel('Pass/Fail (y)')
plt.yticks([0, 1], ['Fail (0)', 'Pass (1)'])
plt.grid(True)
plt.legend()
plt.show()
```

# **Perceptron AND gate**

```python
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
x, y
```

```python
w = np.zeros(2)
b = 0
learning_rate = 0.1
epochs = 10

print("Initial weights: ",w)
print("Initial Bias: ",b)
```

```python
def step_function(z):
  return 1 if z >= 0 else 0
```

```python
for epoch in range(epochs):
  print(f"\nEpoch {epoch + 1}")
  for i in range(len(x)):
    #1. compute weighted sum
    z = np.dot(w, x[i]) + b

    #2. Apply activation function
    y_pred = step_function(z)

    #3. compute error
    error = y[i] - y_pred

    #4. update weight
    w = w + learning_rate * error * x[i]

    #5. update bias
    b = b + learning_rate * error

    print(f"Input: {x[i]}, Predicted: {y_pred}, Actual: {y[i]}, w: {w}, b: {b}" )
```

```python
print("\n Final weight: ",w)
print("\n Final bias: ",b)
```

```python
print("Testing AND gate")
for i in range(len(x)):
  z = np.dot(w, x[i]) + b
  prediction = step_function(z)

  print("Input: ",x[i], "Prediction: ", prediction)
```

```python
import matplotlib.pyplot as plt

# Create a meshgrid to plot the decision boundary
# The decision boundary is defined by w[0]*x1 + w[1]*x2 + b = 0
# So, x2 = (-w[0]*x1 - b) / w[1]
x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))

z_boundary = w[0] * xx + w[1] * yy + b
decision_boundary = np.vectorize(step_function)(z_boundary)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, decision_boundary, alpha=0.4, cmap=plt.cm.coolwarm)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.coolwarm, s=100, edgecolors='k')

plt.title('Perceptron Decision Boundary for AND Gate')
plt.xlabel('Input 1 (x1)')
plt.ylabel('Input 2 (x2)')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks([0, 1])
plt.yticks([0, 1])
plt.grid(True)
plt.show()
```
